```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in os.listdir('.'):
    if i.endswith('.npy'):
        print('File: ', i)
        ar = np.load(str(i))
        plt.imshow(ar, aspect = 'auto')
        plt.show()
```

```python
a = np.load('GBT_A.dm348.8.npy')
b = np.load('GBT_B.dm348.8.npy')
c = np.load('GBT_C.dm348.8.npy')
d = np.load('GBT_D.dm348.8.npy')
e = np.load('GBT_E.dm348.8.npy')
f = np.load('GBT_F.dm348.8.npy')
g = np.load('GBT_G.dm348.8.npy')
```

```python
sub = np.nanmean(a.reshape(-1, 4, a.shape[1]), axis=1)
fig = plt.figure(figsize = (10, 5))
ax = fig.add_subplot(121)
plt.imshow(sub[:,:], aspect = 'auto')
plt.title('A')
ax1 = fig.add_subplot(122)
plt.imshow(sub[:, 150:200], aspect = 'auto')
plt.title('A')
fig.savefig('A.png')
```
